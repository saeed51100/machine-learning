# matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.plot([1, 2, 3, 4], [1, 4, 5, 2])
plt.ylabel("funny numbers")
```

```python
plt.plot([1, 2, 3, 4], [1, 4, 5, 2], "ro")
plt.ylabel("funny numbers")
```

```python
t = np.arange(0, 5, 0.2)
plt.plot(t, t, 'r-.', t, t**2, 'bs', t, t**3, 'g^')
```

```python
data = {
    'a': np.arange(500),
    'b': np.random.randint(0, 50, 50),
    'c': np.random.randn(500)
    }
```
